```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
ax = sns.countplot(x = 'Gender', data = df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen = df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x='Gender' , y='Amount' ,data=sales_gen)
```

```python
sales_gen = df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x='Amount' , y='Gender' ,data=sales_gen)   #wrong method
```

```python
ax = sns.countplot(data = df,x = 'Age Group' , hue = 'Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#total numbers of orders from 10 states
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(20)
sns.set(rc={'figure.figsize':(25,15)})
sns.barplot(data = sales_state, x ='State' , y='Orders')
```

```python
#total amount/sales from top 10 states
sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(16)
sns.set(rc={'figure.figsize':(20,10)})
sns.barplot(data = sales_state, x ='State' , y='Amount')
```

```python
ax = sns.countplot(data = df, x = 'Marital_Status')
sns.set(rc={'figure.figsize':(7,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```
